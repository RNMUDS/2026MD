# 有意差検定（対応のあるt検定）

```python
import pandas as pd
from scipy import stats

# CSVデータ読み込み
df = pd.read_csv('data.csv')
print('データ:')
display(df)
print(f'\nサンプル数: {len(df)}')
```

```python
# 基本統計量
print('=== 基本統計量 ===')
print(df.describe())
print(f'\nA の平均: {df["A"].mean():.3f} (SD: {df["A"].std():.3f})')
print(f'B の平均: {df["B"].mean():.3f} (SD: {df["B"].std():.3f})')
```

```python
# 正規性の検定（Shapiro-Wilk検定）
print('=== 正規性の検定（Shapiro-Wilk） ===')
stat_a, p_a = stats.shapiro(df['A'])
stat_b, p_b = stats.shapiro(df['B'])
print(f'A: W={stat_a:.4f}, p={p_a:.4f} {"→ 正規性あり" if p_a > 0.05 else "→ 正規性なし"}')
print(f'B: W={stat_b:.4f}, p={p_b:.4f} {"→ 正規性あり" if p_b > 0.05 else "→ 正規性なし"}')
```

```python
# 対応のあるt検定
print('=== 対応のあるt検定 ===')
t_stat, p_value = stats.ttest_rel(df['A'], df['B'])
print(f't値: {t_stat:.4f}')
print(f'p値: {p_value:.4f}')
print()
alpha = 0.05
if p_value < alpha:
    print(f'p={p_value:.4f} < {alpha} → AとBの間に有意差あり')
else:
    print(f'p={p_value:.4f} >= {alpha} → AとBの間に有意差なし')
```

```python
# （参考）対応のないt検定・Wilcoxon符号順位検定
print('=== 対応のないt検定（参考） ===')
t2, p2 = stats.ttest_ind(df['A'], df['B'])
print(f't値: {t2:.4f}, p値: {p2:.4f}')

print('\n=== Wilcoxon符号順位検定（ノンパラメトリック、参考） ===')
w_stat, w_p = stats.wilcoxon(df['A'], df['B'])
print(f'W={w_stat:.4f}, p値: {w_p:.4f}')
```
